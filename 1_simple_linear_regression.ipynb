## üìö Import Required Libraries

```python
# Import necessary Python libraries for data handling, visualization, and evaluation
import pandas as pd
import matplotlib.pyplot as plt

# Metrics to evaluate model performance
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error

# Tool to split dataset into training and test sets
from sklearn.model_selection import train_test_split

# Model for prediction
from sklearn.linear_model import LinearRegression
```

## üìÇ Load the Dataset

```python
df=pd.read_csv("SalaryPrediction.csv")
df
```

## üßÆ Splitting Features and Target Variable


```python
# üéØ Define the input feature (Years of Experience) and target variable (Salary)
X = df[["Experience Years"]]  # Features (must be in 2D format)
y = df["Salary"]              # Target variable
```

## Spliting Dataset

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.2,random_state=42)
```

## ü§ñ Fitting the Linear Regression Model to the Dataset

```python
model=LinearRegression()
model.fit(X_train,y_train)

```

## üì§ Making Predictions on Test Data


```python
# üìå Use the trained model to make predictions on the test data
y_pred=model.predict(X_test)
y_pred
```

## üìä Model Evaluation Metrics


```python
# üìè Evaluate the model's performance using different metrics
r2 = r2_score(y_test, y_pred)  # Coefficient of determination
mae = mean_absolute_error(y_test, y_pred)  # Mean Absolute Error
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
rmse = root_mean_squared_error(y_test, y_pred)  # Root Mean Squared Error

# üñ®Ô∏è Display the evaluation results
print("R¬≤ Score:", r2)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)

```

## üìà Visualizing Actual vs Predicted Salaries


```python
# üìä Scatter plot of actual test data
plt.scatter(X_test, y_test, color="blue", label="Actual Data")

# üî¥ Line plot of predicted salaries
plt.plot(X_test, y_pred, color="red", label="Predicted")

# üè∑Ô∏è Add title and axis labels
plt.title("Salary Based on Experience")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.legend()  # Show legend for clarity
plt.show()

```
